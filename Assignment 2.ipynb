# Import Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Dataset Loading

```python
data = pd.read_csv('Annual_Surface_Temperature_Change.csv')
```

```python
print(data.columns)
```

# Data Preprocessing

```python
print(data.isnull().sum())
```

```python
data.drop('CTS_Code', axis=1, inplace=True)
```

```python
data.dropna()
```

# Visualizations

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Country'], data['F1961'])
plt.xlabel('Country')
plt.ylabel('1961')
plt.title('Surface Temperature Change on Country')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Temperature Indicators')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=data['F2022'])
plt.ylabel('Temperature')
plt.title('Distribution of Temperature in 2022')
plt.show()
```

```python
mean_co2_per_year = data.groupby('F1961')['F2022'].mean()
plt.figure(figsize=(10, 6))
mean_co2_per_year.plot(kind='bar', color='skyblue')
plt.xlabel('2022')
plt.ylabel('1961')
plt.title('Temperature Change Over the Year')
plt.show()
```

```python
x_values = data['F1961']
y_values = data['F1962']
plt.scatter(x_values, y_values)
plt.xlabel('1961')
plt.ylabel('1962')
plt.title('Scatter Plot of 1961 vs 1962')
plt.show()
```
